You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.

 

Example 1:


Input: height = [1,8,6,2,5,4,8,3,7]
Output: 49
Explanation: The above vertical lines are represented by array [1,8,6,2,5,4,8,3,7]. In this case, the max area of water (blue section) the container can contain is 49.
Example 2:

Input: height = [1,1]
Output: 1

```python
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        
        #track lp and rp and the width as well

        left = 0
        right = len(height)-1
        spaces = len(height)-1
        max_area = 0                        #used a list initially to append each area and selected the max. But optimized it
                                            #to meet the space O(1)

        
        while height and right > left:
            if height[left] < height[right]:            #when left value < rv, update current_area
                current_area = height[left]*spaces  
                max_area = max(max_area, current_area)      #get max of current_area and max_area
                left += 1

            elif height[left] >= height[right]:
                current_area = height[right]*spaces
                max_area = max(current_area, max_area)
                right -= 1
            spaces -= 1

        return max_area

```
