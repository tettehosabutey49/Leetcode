206. Reverse Linked List

Input: head = [1,2,3,4,5]
Output: [5,4,3,2,1]

```python
# Definition for singly-linked list.
#class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def reverseList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        #curr                     #prev  
        "1 -> 2 -> 3 -> 4 -> 5 -> None"
        "None <- 1  2 -> 3 -> 4 --> 5 -> None"
        "5 -> 4 -> 3 --> 2 -> 1 -> None"

        prev = None
        curr = head
        #iterate through the list
        while curr != None:
          #storing the next node (so we dont lose it after changing the first node's pointer)
          next_node = curr.next    #2
          
          #reversing the pointer
          curr.next = prev         # 1 >> None

          #updating prev and curr for next iteration
          prev = curr
          curr = next_node
        
        return prev 


      
```

Given head, the head of a linked list, determine if the linked list has a cycle in it.

There is a cycle in a linked list if there is some node in the list that can be reached again by continuously following the next pointer. Internally, pos is used to denote the index of the node that tail's next pointer is connected to. Note that pos is not passed as a parameter.

Return true if there is a cycle in the linked list. Otherwise, return false.

 

Example 1:


Input: head = [3,2,0,-4], pos = 1
Output: true
Explanation: There is a cycle in the linked list, where the tail connects to the 1st node (0-indexed).
Example 2:


Input: head = [1,2], pos = 0
Output: true
Explanation: There is a cycle in the linked list, where the tail connects to the 0th node.
Example 3:


Input: head = [1], pos = -1
Output: false
Explanation: There is no cycle in the linked list.

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """
        "3-> 2 -> 0 -> -4 -> 2 -> 0 -4"

        #check edge case first before initializing anything
        if head is None or head.next is None:    #if we have just one or zero entry in the list, it obviously cant have a loop
            return False 

        slow = head                 #moves one step at a time
        fast = head.next            #moves two steps at a time

        while slow is not None and fast is not None:
            if  slow == fast:
                return True
            else:
                slow = slow.next
                fast = fast.next.next
        
        return False
        
```

Time complexity: O(n)     #we iterated through the list
space complexity: O(1)    #we only used two pointors, double and single

You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

 

Example 1:


Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]
Example 2:

Input: list1 = [], list2 = []
Output: []
Example 3:

Input: list1 = [], list2 = [0]
Output: [0]
 

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        "h1           h2"
        "1,2,4"       "1->3,4"
        
        "dummy -> 1 -> 1 -> 2-> 3" 
        "        cur"

        dummy = ListNode()                #dummy node
        current = dummy

        head1 = list1
        head2 = list2

        #catch edge cases
        if list1 is None:     
            return list2

        elif list2 is None:
            return list1 

        while head1 and head2:                      
            if head1.val >= head2.val:
                current.next = head2                #current node should point toward head2 since head2 is smaller
                head2 = head2.next                  #update head2 to the next val to compare with head1 in next iteration
            
            else:
                current.next = head1
                head1 = head1.next
            
            current = current.next          #move current to the next val

        if head1:                            #if there is still items in head1, then next nodes should point the item in head1
            current.next = head1
        
        if head2:
            current.next = head2

        return dummy.next


        
```

-	Time O(n+m)  you traverse each element once in both lists
-	       space O(1)
